```python
semiprime_orders = {}
for order in range(10):
    print('processing order', order)
    semiprimes = [pair[0] * pair[1] for pair in semiprime_factorization.semi_top(order)]
    for semiprime in semiprimes:
        semiprime_orders[semiprime] = order
```

```python
semiprimes = []
orders = []
for semiprime, order in sorted(semiprime_orders.items()):
    semiprimes.append(semiprime)
    orders.append(order)
```
